```python
import pandas as pd
import os

# List of unemployment and population CSV files with their corresponding filenames
unemp_files = ['Citrus_unemp.csv', 'Hernando_unemp.csv', 'Hillsborough_unemp.csv', 'Manatee_unemp.csv', 'Pasco_unemp.csv', 'Pinellas_unemp.csv', 'Polk_unemp.csv', 'Sarasota_unemp.csv']
pop_files = ['Citrus_pop.csv', 'Hernando_pop.csv', 'Hillsborough_pop.csv', 'Manatee_pop.csv', 'Pasco_pop.csv', 'Pinellas_pop.csv', 'Polk_pop.csv', 'Sarasota_pop.csv']

# Corresponding county names for the files
county_names = ['Citrus', 'Hernando', 'Hillsborough', 'Manatee', 'Pasco', 'Pinellas', 'Polk', 'Sarasota']

# Initialize lists to store DataFrames
unemp_dfs = []
pop_dfs = []

# Load unemployment data
for file, county in zip(unemp_files, county_names):
    unemp_df = pd.read_csv(file)
    unemp_df.drop(columns=['realtime_start', 'realtime_end'], inplace=True)
    unemp_df['date'] = pd.to_datetime(unemp_df['date'])  # Convert 'date' to datetime
    unemp_df = unemp_df[(unemp_df['date'] >= '1990-01-01') & (unemp_df['date'] <= '2023-12-31')]  # Filter by date range
    unemp_df.rename(columns={'value': county, 'date': 'Date'}, inplace=True)  # Rename 'value' column to county name and 'date' to 'Date'
    unemp_dfs.append(unemp_df)

# Load population data
for file, county in zip(pop_files, county_names):
    pop_df = pd.read_csv(file)
    pop_df.drop(columns=['realtime_start', 'realtime_end'], inplace=True)
    pop_df['date'] = pd.to_datetime(pop_df['date'])  # Convert 'date' to datetime
    pop_df = pop_df[(pop_df['date'] >= '1990-01-01') & (pop_df['date'] <= '2023-12-31')]  # Filter by date range
    pop_df.rename(columns={'value': county, 'date': 'Date'}, inplace=True)  # Rename 'value' column to county name and 'date' to 'Date'
    pop_dfs.append(pop_df)

# Concatenate unemployment data on 'Date' column
unemployment_df = pd.concat(unemp_dfs, axis=1).loc[:, ~pd.concat(unemp_dfs, axis=1).columns.duplicated()]
unemployment_df = unemployment_df.loc[:, ~unemployment_df.columns.duplicated()]  # Remove duplicate Date columns

# Concatenate population data on 'Date' column
population_df = pd.concat(pop_dfs, axis=1).loc[:, ~pd.concat(pop_dfs, axis=1).columns.duplicated()]
population_df = population_df.loc[:, ~population_df.columns.duplicated()]  # Remove duplicate Date columns

# Display first few rows of both DataFrames
print("Unemployment DataFrame:")
print(unemployment_df.head())

print("\nPopulation DataFrame:")
print(population_df.head())
```

```python
import pandas as pd

unemployment_df['Date'] = pd.to_datetime(unemployment_df['Date'])

# Set 'Date' as the index
unemployment_df.set_index('Date', inplace=True)

# Resample to yearly frequency and calculate the mean of monthly values for each year
yearly_unemployment_df = unemployment_df.resample('Y').mean()

# Reset index to have 'Date' as a column again
yearly_unemployment_df.reset_index(inplace=True)

# Format 'Date' column to yyyy-01-01
yearly_unemployment_df['Date'] = yearly_unemployment_df['Date'].dt.year.astype(str) + '-01-01'

# Convert the 'Date' column to datetime format
yearly_unemployment_df['Date'] = pd.to_datetime(yearly_unemployment_df['Date'])

# Display the first few rows of the yearly unemployment DataFrame
print("Yearly Unemployment DataFrame:")
print(yearly_unemployment_df.head())
```

```python
# Set 'Date' as index
yearly_unemployment_df.set_index('Date', inplace=True)
population_df.set_index('Date', inplace=True)

# Calculate yearly weighted unemployment rate
def calculate_weighted_unemployment_rate(unemployment_df, population_df):
    weighted_rates = []
    for date in unemployment_df.index:
        unemployment_rates = unemployment_df.loc[date]
        populations = population_df.loc[date]
        total_population = populations.sum()
        weighted_rate = (unemployment_rates * populations).sum() / total_population
        weighted_rates.append(weighted_rate)
    return pd.DataFrame({
        'Date': unemployment_df.index,
        'Weighted_Unemployment_Rate': weighted_rates
    })

# Calculate and create new DataFrame
df_weighted_unemployment = calculate_weighted_unemployment_rate(yearly_unemployment_df, population_df)

# Display the new DataFrame
print(df_weighted_unemployment)
```
